### 1. Bigram (two-word phrases0 frequency)
Instead of single words, look at common pairs of words (like “net sales”, “profit increased”).

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_top_ngrams(corpus, n=2, top=10):
    vec = CountVectorizer(ngram_range=(n,n), stop_words="english").fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    return sorted(words_freq, key = lambda x: x[1], reverse=True)[:top]

# Top bigrams overall
top_bigrams = get_top_ngrams(financial_news['text'].astype(str), n=2, top=15)
print("Top bigrams:", top_bigrams)
```

Filter by sentiment class:

```python
top_bigrams_pos = get_top_ngrams(financial_news[financial_news['sentiment']=="positive"]['text'].astype(str), n=2, top=10)
print("Positive bigrams:", top_bigrams_pos)
```
